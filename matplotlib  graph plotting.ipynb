```python
x=np.arange(4)
car1=[40,17,28,12]
car2=[34,21,24,15]
```

```python
plt.bar(x,car1,width=0.3,label='car1',tick_label=[2017,2018,2019,2020])
plt.bar(x+0.3,car2,width=0.3,label='car2')
plt.xlabel('years')
plt.ylabel('sales in million dollar')
plt.title('cars price comparison')
plt.legend
plt.xlim(-1,5)
plt.ylim(-5,50)#x+0.3(for remove overlapping of bar graphs)
```

```python
#Histograms....plt.hist()........creating dummy data using normal distribution
```

```python
X_snd=np.random.randn(500)
```

```python
math_mu=70
math_sig=8
math_marks=np.round(X_snd*math_sig+math_mu)

phy_mu=50
phy_sig=10
phy_marks=np.round(X_snd*phy_sig + phy_mu)
```

```python
plt.hist(math_marks,bins=20)
plt.hist(phy_marks,alpha=0.7,bins=20)
plt.show()
```
